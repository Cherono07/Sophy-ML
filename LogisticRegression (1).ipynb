## OPTION1: Imputing missing data

```python
# this helps plot Age numbers of pple without values (NaN= those who have NO values)
# if a person is in class 1, they are automatically given the value 37 because it is the average mean of that class
# if a person is in class 2, they are automatically given the value 29 because it is the average mean of that class
# if a person is in class 3, they are automatically given the value 24 because it is the average mean of that class

def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]

  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 29
    else:
      return 24
  else:
    return Age
```

```python
# You're selecting two columns: Age and Pclass.....-> The Age column has missing values that you want to fill in (impute),
                                                 # -> and you're using the Pclass column to help decide how to fill those missing values.
# AXIS PARAMETER controls whether the function is applied across rows or across columns.
                 #->  axis=0 (default): The function is applied COLUMN-WISE, meaning it operates on each column of the DataFrame.
                 #-> axis=1: The function is applied ROW-WISE, meaning it operates on each row of the DataFrame.

#THEREFORE....axis=1 specifies that the function should be applied to each row (not column-wise).
#train['Age']= ....column we want to change

train['Age']= train[['Age', 'Pclass']].apply(impute_age, axis=1) # gives the NaN values
```

```python
#now in the heatmap the Age column has now no missing values

sns.heatmap(train.isnull(),yticklabels=False,cbar=True,cmap="viridis")
```

## Categorical data into dummy variables
Converting all our categorical data to dummy variables using Pandas

```python
# provides summary of our data

train.info()
```

```python
#converts the the sex data to binary data.....true/false || 0/1
# so if male =true and  if not male(if female)==false

sex=pd.get_dummies(train['Sex'],drop_first=True)
sex
```

```python
embark= pd.get_dummies(train['Embarked'], drop_first=True)
embark
```

```python
# train.drop()......This is a method used to remove rows or columns from a DataFrame.
# we drop the columns.....['Sex', 'Embarked', 'Name', 'Ticket']
       # -> we drop the SEX and EMBARKED inorder to add the dummy data we invoked above and get rid of the previous categorical data
       # -> we drop the NAME because it is not a feature.
       # -> we drop the TICKET because it is NOT important for us to determine the answers
# axis=1......means you're dropping columns (not rows).
# inplace=True.....This modifies the original DataFrame (train) directly, meaning the columns will be removed from the train DataFrame itself.
      # -> If you set inplace=False (which is the default), it would return a new DataFrame with the columns removed, but the original train DataFrame would remain unchanged.

train.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
train
```

```python
# pd.concat() is a function from pandas used to concatenate (join or merge) multiple DataFrames or Series along a specified axis (rows or columns).
# [train, sex, embark] is a list of the DataFrames (or Series) you want to concatenate. This line will join:
# axis=1....specifies that the concatenation should happen along the columns (side-by-side).
       #-> If you used axis=0, the concatenation would happen along the rows (stacking vertically).
       # -> So, the sex and embark columns will be added to the train DataFrame as new columns.


train=pd.concat([train, sex, embark], axis=1)
```

## Spliting x and y as well as the train and test

```python
from sklearn.model_selection import train_test_split
```

```python
#(train.drop('Survived', axis=1)--> x axis, we drop survived because it is a label
#train['Survived']-->y axis
#test_size=0.20--> 20% used for testing and thus 80% used for training the model
#Random state is used to ensure that when you split your dataset, you get the same split every time you run the code.


X_train, X_test, y_train, y_test= train_test_split(train.drop('Survived', axis=1), train['Survived'], test_size=0.20, random_state=101)
```

## Building model

## Evaluating Model[train vs test] using "metrics"
-  auc==> The Area Under the Curve of the Receiver Operating Characteristic (ROC).   It tells you how well your model distinguishes between classes (higher is better).
- accuracy_score--> A measure of how many predictions were correct compared to the total number of predictions.
- classification_report--> A detailed report that provides precision, recall, f1-score, and support for each class.
- log_loss--> This measures how far your predicted probabilities are from the actual values. LOSS FUNCTION!!.
- confusion_matrix==> A table that shows if the model predicted 0 or 1 correctly or incorrectly.
-------------------------------------------------------------------------------
- 1. Precision: How many of the things the model guessed as "positive" (e.g., survived) were actually correct?
        #-> For example, if the model predicted 10 people survived, and 8 of those 10 were correct, the precision is 80%.

-  2. Recall: How many of the actual "positives" (e.g., people who survived) did the model manage to catch?

- 3. F1-Score: A balanced number that combines precision and recall.
     #-> finding a middle ground between how well the model was at guessing and how many real positives it found.

- 4. Support: The number of real instances for each class (e.g., how many people actually survived vs. how many did not survive in the test data).

```python
from sklearn.metrics import auc, accuracy_score, classification_report, log_loss, confusion_matrix
print(classification_report(y_test, predictions))
```

```python
# the leading diagonals in the matrix is the one the model got right

print(sns.heatmap(confusion_matrix(predictions, y_test), annot=True))
```
